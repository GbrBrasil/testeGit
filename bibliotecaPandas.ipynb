```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel('/content/varejo.xlsx')
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP','Aplicativo')
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby('estado').idcompra.nunique()
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ','_')
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.info()
```

```python
vendas[vendas['estado'].isnull()]
```

```python
vendasSemNulos = vendas.dropna()
vendasSemNulos.info()
```

```python
vendas['estado'].fillna('MS', inplace=True)
```

```python
vendas.info()
```

```python
mediaPreco = vendas['Preço'].mean()

vendas['Preço'].fillna(mediaPreco, inplace = True)
```

```python
mediaPreco
```

```python
vendas.info()
```

```python
vendas[['Preço','Preço_com_frete']].describe()
```

```python
precoErrado = vendas.query('Preço > Preço_com_frete')

vendasCorreto = vendas.query('Preço < Preço_com_frete')
```

```python
vendasCorreto.describe()
```

```python
vendasCorreto.head()
```

```python
vendasCorreto.query('Nome_Departamento == "Esporte_e_Lazer" and estado == "SP"').describe()
```

```python
vendasCorreto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
round(vendasCorreto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending = False).reset_index(),2)
```

```python
vendasCorreto.groupby('Data').idcompra.nunique().sort_values(ascending=False)
```

```python
vendasCorreto['mes'] = vendasCorreto['Data'].dt.month
vendasCorreto.head()
```

```python
vendasCorreto.groupby('mes').idcompra.nunique().sort_values(ascending=False)
```

```python
cliente = pd.read_excel('/content/cliente_varejo.xlsx')
cliente.head()
```

```python
cliente[['idade','renda']].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({"renda":"float"})
cliente.info()
```

```python
vendasCliente = vendasCorreto.merge(cliente, how = 'left', on = 'cliente_Log')
vendasCliente.head()
```

```python
aggIdCanalRenda = round(vendasCliente.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending=False).reset_index(),2)
aggIdCanalRenda
```

```python
aggIdadeBandeira = round(vendasCliente.groupby('bandeira')['idade'].agg('mean').sort_values(ascending = False).reset_index(),0)
aggIdadeBandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(aggIdadeBandeira['bandeira'], aggIdadeBandeira['idade'],color='green')
plt.ylabel('Média de idade')
plt.title('Idade média por bandeira', loc='center',fontsize=18,fontweight='bold',color='black');
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(aggIdCanalRenda['idcanalvenda'], aggIdCanalRenda['renda'],color='green');
plt.ylabel('Renda média')
plt.title('Renda média por canal de venda',loc='center', fontsize=18,fontweight='bold',color='black');
```

```python
vendasPorData = vendasCorreto.groupby('Data').idcompra.nunique().reset_index()
vendasPorData
```

```python
fig = plt.figure(figsize=(20,4))

plt.xlabel('Data da venda')
plt.ylabel("Quantidade de vendas")
x = vendasPorData['Data']
y = vendasPorData['idcompra']

plt.plot(x,y, color = "green",linewidth = '2')
```

```python
aggDeptPreco = round(vendasCorreto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending = False).reset_index(),2)
aggDeptPreco

fig = plt.figure(figsize=(50,30))

plt.bar(aggDeptPreco['Nome_Departamento'], aggDeptPreco['Preço_com_frete'],color='green');
plt.ylabel('Preço médio')
plt.title('Preço médio por departamento',loc='center', fontsize=18,fontweight='bold',color='black');
```
